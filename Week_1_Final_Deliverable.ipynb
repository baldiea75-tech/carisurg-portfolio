```python
week1_readme = """# Week 1 — Interim Deliverable

## Project Title
Development of an Interpretable Digital Triage System for Dynamic Risk Assessment in Emergency Care

## Author
Ashi Baldie

## Programme
CariSurg MedTech Pathways 2026
Mercer General Hospital | Clinical AI & Innovation Unit

---

## What This Week Covers

Week 1 shifts from data cleaning (Week 0) into research and system design.
The interim deliverable is a structured research document that defines the
clinical problem, reviews the existing literature, and proposes a solution.

---

## Deliverable

| File | Description |
|------|-------------|
| `Week1_Interim_Deliverable.pdf` | Full interim research document |

---

## Problem Statement

Emergency department overcrowding delays care for critically ill patients
because traditional triage relies on static, point-in-time assessments
that fail to capture real-time physiological deterioration in the waiting room.
Current AI triage solutions operate as uninterpretable black boxes, hindering
clinical adoption and alignment with established hospital protocols.

---

## Proposed Solution

A transparent digital triaging tool for real-time decision support at
Mercer General Hospital ED. Key features:

- Processes structured vital signs (GCS, SBP, DBP, MAP, pulse, Temp, RR, FiO2)
- Incorporates unstructured free-text nurse notes via NLP
- Updates risk level dynamically as new patient data arrives (not just on arrival)
- Aligns outputs with NEWS2 and SIRS frameworks staff already know
- Displays interpretable explanations for every risk decision (no black box)

---

## Key Literature

| Citation | Relevance |
|----------|-----------|
| Tyler et al. (2024) | ML models match or exceed human triage accuracy |
| Da'Costa et al. (2025) | Black-box problem blocks clinical adoption |
| Sitthiprawiat et al. (2025) | XGBoost outperforms CTAS (AUROC 0.917 vs 0.882) |
| Stewart et al. (2023) | NLP on free-text triage notes improves prediction |
| Boulitsakis Logothetis et al. (2023) | Interpretable ML outperforms NEWS2 framework |
| Liu et al. (2022) | Dynamic interpretable scoring across ED stay |
| Araouchi & Adda (2024) | Multimodal fusion of structured + unstructured data |

---

## Status
- [x] Problem statement defined
- [x] Literature review complete (7 peer-reviewed sources, 2022-2025)
- [x] Proposed solution documented
- [ ] Week 2: Pseudocode and system architecture
- [ ] Week 12: Live pilot deployment at Mercer General ED

---

## Branch
`week1-interim-deliverable`
"""

with open('week1/README.md', 'w') as f:
    f.write(week1_readme)

print("Week 1 README created.")
```

```python
import json

notebook_path = '/content/carisurg-portfolio/week1/Week1_Final_Deliverable.ipynb'

with open(notebook_path, 'r') as f:
    nb = json.load(f)

for cell in nb['cells']:
    if cell['cell_type'] == 'code':
        # Clear all outputs
        cell['outputs'] = []
        cell['execution_count'] = None
        # Remove any line containing a token or remote set-url command
        cleaned = []
        for line in cell['source']:
            if 'ghp_' in line or 'git remote set-url' in line:
                cleaned.append('# [credential removed]\n')
            else:
                cleaned.append(line)
        cell['source'] = cleaned

with open(notebook_path, 'w') as f:
    json.dump(nb, f, indent=1)

print("Notebook cleaned.")
```
